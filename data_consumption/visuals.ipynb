```python
import pandas as pd

df = pd.read_parquet("./data.parquet")

df
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-v0_8-whitegrid")
plot_df = df.sort_values("State_Name")
x = np.arange(len(plot_df))

bmi_colors = plt.cm.viridis(np.linspace(0.18, 0.88, len(plot_df)))
count_colors = plt.cm.plasma(np.linspace(0.18, 0.88, len(plot_df)))

fig, ax = plt.subplots(figsize=(18, 7))
ax.bar(x, plot_df["bmi_mean"], color=bmi_colors, edgecolor="white", linewidth=0.7)
ax.set_title("Average BMI by State", fontsize=16, weight="bold", pad=14)
ax.set_xlabel("State")
ax.set_ylabel("BMI mean")
ax.set_xticks(x)
ax.set_xticklabels(plot_df["State_Name"], rotation=45, ha="right")
ax.grid(axis="y", linestyle="--", alpha=0.35)
ax.grid(axis="x", visible=False)
ax.spines[["top", "right"]].set_visible(False)
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(18, 7))
ax.bar(x, plot_df["record_count"], color=count_colors, edgecolor="white", linewidth=0.7)
ax.set_title("Record Count by State", fontsize=16, weight="bold", pad=14)
ax.set_xlabel("State")
ax.set_ylabel("Record count")
ax.set_xticks(x)
ax.set_xticklabels(plot_df["State_Name"], rotation=45, ha="right")
ax.grid(axis="y", linestyle="--", alpha=0.35)
ax.grid(axis="x", visible=False)
ax.spines[["top", "right"]].set_visible(False)
plt.tight_layout()
plt.show()
```
